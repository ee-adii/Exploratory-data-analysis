```python
apple  = yf.Ticker('AAPL')
```

```python
apple_data =apple.info
apple_data
```

```python
df = pd.DataFrame([apple_data])
df
```

```python
apple_stocks = apple.history(period="MAX")
apple_stocks  = pd.DataFrame(apple_stocks)
apple_stocks
```

## working with dates 

```python
dates = apple_stocks.index
apple_stocks.index = [i for i in range(10346)]

```

```python
help(df.insert)
```

```python
apple_stocks
```

```python
apple_stocks.insert(loc = 0 , column = 'Date' , value  = dates)
```

```python
apple_stocks
```

```python
apple_stocks.Date = pd.to_datetime(apple_stocks.Date)
```

```python
## adding year , week , Month column in df by DateTimeIndex method

apple_stocks["Year"]  = pd.DatetimeIndex(apple_stocks.Date).year
apple_stocks["Month"]  =pd.DatetimeIndex(apple_stocks.Date).month
apple_stocks["Weekday"] = pd.DatetimeIndex(apple_stocks.Date).weekday
apple_stocks["day"] = pd.DatetimeIndex(apple_stocks.Date).day
```

```python
apple_stocks
```

```python
apple_stocks_dec = apple_stocks[apple_stocks.Month == 12]
apple_stocks_dec.tail(5)
```

## Grouping and Aggregation

#### Somtimes we might want to summarize the data from month , weeks , years
#### to do so , we have grupby method to aggregate the data from given months 

```python
apple_stocks_monthwise = apple_stocks.groupby('Month')
## this will aggregate data from all the months
```

```python
apple_stocks_monthwise[["Open" , "Low" , 'Volume']].max()  ## max value from mention columns
```

```python
apple_stocks_monthwise[["Dividends"]].sum()
```
