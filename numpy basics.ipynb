## Numpy axis
### for 1D only one axis i.e axis =0
### for 2D two axis i.e. axis =1, axis =0     (0 for rows and 1 for coloums)

```python
ll = np.array([[2,3,4],[5,2,1],[84,2,8]])
```

```python
ll.sum(axis = 0)       ## sum across rows
```

```python
ll.sum(axis = 1)  ### sum across coloums
```

```python
ll.T          ### gives the transpose of numpy array
```

```python
ll.ndim      ### gives the dimension of numpy array
```

```python
### argmin() and argmax gives index of minimum and maximum
ar = np.array([3,4,5,645,-2,78])
ar.argmax()    
ar.argmin()        ## you can pass axis to find max or min in a row or coloumn
```

```python
## argsort()     gives the array of index of sorted element 
ar.argsort()
```

```python
### reshape((a,b)) gives array of shape (a,b)    
```

```python
ar.reshape((6,1))
```

```python
ar.reshape((6,))
```

```python
### np.where(condition) gives the location of elements satisfy the condition

np.where(ll>2)
```

```python
### np.count_nonzero("list")   gives numbers of non-zero

np.count_nonzero(ar)
```
